```python
import pandas as pd
from google.colab import files

# =========================
# 1. UPLOAD
# =========================

print("Faça upload do arquivo:")
uploaded = files.upload()

arquivo = list(uploaded.keys())[0]
df = pd.read_excel(arquivo)

# =========================
# 2. LIMPEZA BÁSICA
# =========================

df.columns = df.columns.str.strip()

# =========================
# 3. DEFINIÇÃO DAS COLUNAS (EXATAS DO IBGE)
# =========================

col_ano = "Ano"
col_municipio = "Nome do Município"

col_pib = """Produto Interno Bruto,
a preços correntes
(R$ 1.000)"""

col_pib_per_capita = """Produto Interno Bruto per capita,
a preços correntes
(R$ 1,00)"""

# =========================
# 4. VALIDAÇÃO
# =========================

for col in [col_ano, col_municipio, col_pib, col_pib_per_capita]:
    if col not in df.columns:
        print("\nColunas disponíveis:")
        for c in df.columns:
            print(repr(c))
        raise ValueError(f"Coluna não encontrada: {col}")

# =========================
# 5. FILTRO DE ANOS
# =========================

df = df[df[col_ano].between(2016, 2021)]

# =========================
# 6. FILTRO SUDOESTE PR
# =========================

municipios_sudoeste = [
    "Pato Branco", "Francisco Beltrão", "Dois Vizinhos", "Palmas",
    "Capanema", "Realeza", "Ampére", "Barracão",
    "Santo Antônio do Sudoeste", "Coronel Vivida",
    "Chopinzinho", "Mangueirinha", "Salto do Lontra",
    "Itapejara d'Oeste", "Marmeleiro", "Nova Prata do Iguaçu"
]

df[col_municipio] = df[col_municipio].str.strip().str.title()
df = df[df[col_municipio].isin(municipios_sudoeste)]

# =========================
# 7. VISUALIZAR DADOS
# =========================

print("\n=== DADOS FILTRADOS ===")
display(df[[col_ano, col_municipio, col_pib, col_pib_per_capita]].head(30))

print(f"\nTotal de municípios: {df[col_municipio].nunique()}")

# =========================
# 8. SALVAR
# =========================

df.to_excel("dados_filtrados.xlsx", index=False)
files.download("dados_filtrados.xlsx")

# =========================
# 9. ANÁLISE
# =========================

ultimo_ano = df[col_ano].max()
df_ultimo = df[df[col_ano] == ultimo_ano]

# Ranking PIB
ranking_pib = df_ultimo.sort_values(by=col_pib, ascending=False)

print("\n=== TOP 10 PIB ===")
display(ranking_pib[[col_municipio, col_pib]].head(10))

# Ranking PIB per capita
ranking_pc = df_ultimo.sort_values(by=col_pib_per_capita, ascending=False)

print("\n=== TOP 10 PIB PER CAPITA ===")
display(ranking_pc[[col_municipio, col_pib_per_capita]].head(10))

# Crescimento
crescimento = (
    df.sort_values(col_ano)
    .groupby(col_municipio)[col_pib]
    .apply(lambda x: (x.iloc[-1] / x.iloc[0] - 1) * 100 if len(x) > 1 else None)
)

print("\n=== CRESCIMENTO (%) ===")
display(crescimento.dropna().sort_values(ascending=False))
```
